```python
import seaborn as sb
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_pickle("test_best_train.p")
```

```python
df
```

### Stripplot with median combing from boxplot

Boxplot references:
https://matplotlib.org/3.1.0/gallery/statistics/boxplot.html

```python
plt.figure(figsize=(5,3), dpi=300)
sb.boxplot(x="method", y="R2", hue="lat", order=["LM", "NN", "SPOC_LM", "SPOC_XGB","XGB"], 
           data=df[df["loc"] == "ECOG"], palette="viridis", 
           showmeans=False, boxprops=dict(alpha=1), showcaps=False, showbox=False, showfliers=False, 
           whiskerprops={'linewidth':0, "zorder":10},
           medianprops=dict(linestyle='-.', linewidth=5, color="gray", alpha=0.7))

    
ax = sb.stripplot(x="method", y="R2", hue="lat", order=["LM", "NN", "SPOC_LM", "SPOC_XGB","XGB"], 
             data=df[df["loc"] == "ECOG"],palette="viridis", dodge=True, s=5)


# When creating the legend, only use the first two elements
# to effectively remove the last two.
handles, labels = ax.get_legend_handles_labels()
l = plt.legend(handles[0:2], labels[0:2], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title("ECOG Test(Best Train) Nested CV Results")
```

### Overlayed Boxplot and Striplot

```python
alpha_box = 0.4
plt.figure(figsize=(5,3), dpi=300)
sb.boxplot(x="method", y="R2", hue="lat", order=["LM", "NN", "SPOC_LM", "SPOC_XGB","XGB"], 
           data=df[df["loc"] == "ECOG"], palette="viridis", 
           showmeans=False, boxprops=dict(alpha=alpha_box), showcaps=True, showbox=True, 
           showfliers=False, notch=False,
           whiskerprops={'linewidth':2, "zorder":10, "alpha":alpha_box},
           capprops={"alpha":alpha_box},
           medianprops=dict(linestyle='-.', linewidth=5, color="gray", alpha=alpha_box))

    
ax = sb.stripplot(x="method", y="R2", hue="lat", order=["LM", "NN", "SPOC_LM", "SPOC_XGB","XGB"], 
             data=df[df["loc"] == "ECOG"],palette="viridis", dodge=True, s=5)


# When creating the legend, only use the first two elements
# to effectively remove the last two.
handles, labels = ax.get_legend_handles_labels()
l = plt.legend(handles[0:2], labels[0:2], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title("ECOG Test(Best Train) Nested CV Results")
```

```python
alpha_box = 0.4
plt.figure(figsize=(5,3), dpi=300)
ax = sb.violinplot(x="method", y="R2", hue="lat", order=["LM", "NN", "SPOC_LM", "SPOC_XGB","XGB"], 
           data=df[df["loc"] == "ECOG"], palette="viridis")
plt.setp(ax.collections, alpha=alpha_box)
    
ax = sb.stripplot(x="method", y="R2", hue="lat", order=["LM", "NN", "SPOC_LM", "SPOC_XGB","XGB"], 
             data=df[df["loc"] == "ECOG"],palette="viridis", dodge=True, s=5)


# When creating the legend, only use the first two elements
# to effectively remove the last two.
handles, labels = ax.get_legend_handles_labels()
l = plt.legend(handles[0:2], labels[0:2], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title("ECOG Test(Best Train) Nested CV Results")
```
