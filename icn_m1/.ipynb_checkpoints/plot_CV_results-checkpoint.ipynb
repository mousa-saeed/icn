```python
def get_best(ECOG=True, Contra=True):
    best_train_gridpoint = np.zeros(17)
    best_train_val = np.zeros(17)
    test_val = np.zeros(17)
    if ECOG is True:
        if Contra is True:
            ind_low = 0; ind_high = 39; 
        else:
            ind_low = 39; ind_high = 78; 
    else: 
        if Contra is True:
            ind_low = 78; ind_high = 86; 
        else:
            ind_low = 86; ind_high = 94


    for patient_idx in range(17):
        best_train_gridpoint[patient_idx] = np.argmax(auc_train[patient_idx, ind_low:ind_high]) + ind_low
        best_train_val[patient_idx] = np.max(auc_train[patient_idx, ind_low:ind_high])
        test_val[patient_idx] = auc_test[patient_idx,  int(best_train_gridpoint[patient_idx])]
    return best_train_gridpoint, best_train_val, test_val
```

```python
def get_weighted_val(ECOG=True, Contra=True, n_gridpoints=3):
    res_out = []; weighted_mean_out = np.zeros(17); grid_points_used = []
    best_train_gridpoint = np.zeros(17)
    best_train_val = np.zeros(17)
    test_val = np.zeros(17)
    if ECOG is True:
        if Contra is True:
            ind_low = 0; ind_high = 39; 
        else:
            ind_low = 39; ind_high = 78; 
    else: 
        if Contra is True:
            ind_low = 78; ind_high = 86; 
        else:
            ind_low = 86; ind_high = 94

    for patient_idx in range(17):
        print('patient: '+ str(patient_idx))
        vals = np.flip(np.sort(auc_train[patient_idx,ind_low:ind_high])[-n_gridpoints:])
        argvals = np.flip(np.argsort(auc_train[patient_idx,ind_low:ind_high])[-n_gridpoints:]) + ind_low
        grid_points_used.append(argvals)
        weights = (vals - 0.5) / np.sum(vals - 0.3)
        print('weights:')
        print(weights)
        start = 0

        if patient_idx < 10:
            subject_id = str('00') + str(patient_idx)
        else:
            subject_id = str('0') + str(patient_idx)

        res_path = '/Users/hi/Documents/workshop_ML/thesis_plots/int_out_downsampled/'+subject_id+'prediction.npy' 
        res = np.load(res_path, allow_pickle=True)
        print('used grid points: ')
        print(argvals)
        for idx, grid_point in enumerate(argvals):
            if weights[idx] <= 0:
                continue
            print(grid_point)
            if start == 0:
                start = 1
                try:
                    out = weights[idx] * res[grid_point]['y_pred_test']
                except:
                    pass
            else:
                try:
                    out += weights[idx] * res[grid_point]['y_pred_test']
                except:
                    print('unterschiedliche anzahl von daten')
                    pass
        res_out.append(out)
        if patient_idx != 16:
            try:
                weighted_mean_out[patient_idx] = roc_auc_score(res[grid_point]['y_test'][100:4000]>0, out[100:4000])
            except:
                pass
        else:
            try:
                weighted_mean_out[patient_idx] = roc_auc_score(res[grid_point]['y_test'][100:4000]>0, out[100:4000])
            except:
                pass
    return res_out, weighted_mean_out, grid_points_used
```

```python
res_out, weighted_mean_out, grid_points_used = get_weighted_val(ECOG=True, \
                                                Contra=True, n_gridpoints=3)
weighted_mean_out_ECOG_con = weighted_mean_out
res_out, weighted_mean_out, grid_points_used = get_weighted_val(ECOG=True, \
                                                Contra=False, n_gridpoints=3)
weighted_mean_out_ECOG_ips = weighted_mean_out
res_out, weighted_mean_out, grid_points_used = get_weighted_val(ECOG=False, \
                                                Contra=True, n_gridpoints=3)
weighted_mean_out_STN_con = weighted_mean_out
res_out, weighted_mean_out, grid_points_used = get_weighted_val(ECOG=False, \
                                                Contra=False, n_gridpoints=3)
weighted_mean_out_STN_ips = weighted_mean_out
```
