#### time frequency transform using digital FIR filter

```python
def t_f_transform(x, sample_rate, f_ranges):
    """
    :param x: given array in form [f_range, time]
    :param sample_rate
    :param f_ranges
    :return: list of filtered stream for all frequency bands
    """

    filtered_x = []
    nyq_rate = sample_rate / 2
    width = 1.0 / nyq_rate
    N, beta = signal.kaiserord(settings.ripple_db, width)
    noise_ = np.arange(settings.line_noise[0], settings.line_noise[1] + 1, 1)

    for f_range in f_ranges:

        cutoff_hz = f_range  #BP Filter range
        f_range = np.arange(f_range[0], f_range[1] + 1, 1)
        taps = signal.firwin(N, np.array(cutoff_hz) / nyq_rate, window = ('kaiser', beta), pass_zero=False, fs=sample_rate)


        if np.intersect1d(noise_, f_range).shape[0] != 0:
            #do line noise filtering
            taps_ = signal.firwin(N, np.array(settings.line_noise) / nyq_rate, window=('kaiser', beta),pass_zero=False, fs=sample_rate)
            x = signal.lfilter(taps_, 1.0, x)

        time_filtered = signal.lfilter(taps, 1.0, x)
        filtered_x.append(time_filtered)
    return np.array(filtered_x)


def transform_channels(bv_raw):
    """
    calculate t-f-transform for every channel
    :param bv_raw: Raw (channel x time) datastream
    :return: t-f transformed array in shape (len(f_ranges), channels, time)
    """
    x_filtered = np.zeros([len(settings.f_ranges), bv_raw.shape[0] - 2, bv_raw.shape[1]])
    for ch in range(bv_raw.shape[0] - 2):
        x_filtered[:, ch, :] = t_f_transform(bv_raw[ch, :], settings.sample_rate, settings.f_ranges)
    return x_filtered

def plot_frequencies(x_filtered_zscored):
    #Frequency bands
    plt.figure(figsize=(15,6), dpi=300)
    for f_idx in range(len(settings.f_ranges)):
        f, t, Sxx = sp.signal.spectrogram(x_filtered_zscored[f_idx,0,:], settings.sample_rate)
        plt.subplot(2,4,f_idx+1)
        plt.pcolormesh(t, f, Sxx); plt.ylim(0,200)
        plt.title('Frequency band: '+str(settings.f_ranges[f_idx])+' Hz')
        plt.xlabel('Time [s]'); plt.ylabel('Frequency [Hz]')
        plt.colorbar()
        plt.clim(0, 0.01)
    plt.tight_layout()
```

```python
def filter_this(x):
    flim = [1, 400]
    sfreq = settings.sample_rate
    x_filt_mne = []
    noise_ = np.arange(settings.line_noise[0], settings.line_noise[1] + 1, 1)

    for f in range(len(settings.f_ranges)):
        
        f_range = settings.f_ranges[f]
        if np.intersect1d(noise_, f_range).shape[0] != 0:
            #do line noise filtering
            mne.filter.notch_filter
        
        l_freq = settings.f_ranges[f][0]
        h_freq = settings.f_ranges[f][1]
        h = mne.filter.create_filter(x, sfreq, l_freq=l_freq, h_freq=h_freq, \
                                     fir_design='firwin', verbose=False, l_trans_bandwidth=2, h_trans_bandwidth=2)
        x_filt_mne.append(np.convolve(h, x, mode='same'))
        # this is the linear->zero phase, causal-to-non-causal conversion / shift
        #x_v16 = x_v16[len(h) // 2:]

        #plot_filter(h, sfreq, freq, gain, 'MNE-Python 0.16 default', flim=flim,
        #            compensate=True)
    return np.array(x_filt_mne)
```

```python
def check_leave_out_grid_points(act_):
    """
    :param act_: array of run files of each patient having interpolated grid points
    :return: a list of grid points which are only occuring in none or one patient
    """
    grid_point_occurance = np.zeros(94)
    for patient_idx in range(16):
        grid_point_occurance[np.where(np.sum(act_[patient_idx], axis=0))[0]] += 1
    grid_points_none = list(np.where(grid_point_occurance == 1)[0]).append(np.where(grid_point_occurance == 0)[0])
    return grid_points_none
```

```python
def check_leave_out_grid_points(act_, load=True):
    """
    :param act_: array of run files of each patient having interpolated grid points
    :return: a list of grid points which are only occuring in none or one patient
    """
    if load is True:
        return np.load('grid_points_none.npy', allow_pickle=True)
    else:
        grid_point_occurance = np.zeros(94)
        for patient_idx in range(16):
            grid_point_occurance[np.where(np.sum(act_[patient_idx], axis=0))[0]] += 1
        grid_points_none = np.where((grid_point_occurance == 0) | (grid_point_occurance == 1))[0]
        np.save('grid_points_none.npy', grid_points_none)
        return grid_points_none
```
